```python
#This codes can be used to compare 2 DNA sequences, transcribe DNA to RNA,and translate RNA to Protein sequence.

#This part codes for the DNA comparision: It will display the position of difference and the difference in nucleotides.

def sequence_comparision(a,b):
    similar=""
    extra=""
    print("total base pair in 1st dna sequence=",len(a))
    print("total base pair in 2nd dna sequence=",len(b))
    
    temp_a=a.upper()
    temp_b=b.upper()
    count=0
    if len(temp_a)>len(temp_b):
        diff=len(temp_a)-len(temp_b)
        for i in range(0,len(temp_b)):
            if temp_a[i]==temp_b[i]:
                similar+=temp_b[i]
                continue
            else:
                count+=1
                print('Differnece of base pair at',(i+1),'position.', '\nBasepair in sequence 1=',temp_a[i],'\nBasepair in sequence 2=',temp_b[i])
        for ex in range (len(temp_b),len(temp_a)):
            extra+=temp_a[ex]
        print("Extra sequence=", extra)
    else:
        for j in range (0,len(temp_a)):
            if temp_a[j]==temp_b[j]:
                similar+=temp_a[j]
                continue
            else:
                count+=1
                print('Differnece of base pair at',(j+1),'position.', '\nBasepair in sequence 1=',temp_a[j],'\nBasepair in sequence 2=',temp_b[j])
        for ex in range (len(temp_a),len(temp_b)):
            extra+=temp_b[ex]
        print("Extra sequence=", extra)
    print("number of differences=", count)
    print("similar sequence=",similar)
print("DNA SEQUENCE COMPARISON")
a_name= input("Enter name of sequence 1:")    
a= input("\nEnter sequence here:")
a_name= input("\nEnter name of sequence 2:") 
b= input("\nEnter sequence here:")
sequence_comparision(a,b)

#This codes to extract the exons
exon_a=""
for chr in a:
    if chr=="a"or chr=="t" or chr=="g" or chr=="c":
        continue
    else:
        exon_a+=chr

exon_b=""
for chrt in b:
    if chrt=="a"or chrt=="t" or chrt=="g" or chrt=="c":
        continue
    else:
        exon_b+=chrt


#This function will calculate the frequency of the 4 nucleotides in both the sequences and then calculate the PCR tempurate for both the sequences.
def percentage_basepairs(sequence):
    total_base_pairs= len(sequence)
    adenine=0
    guanine=0
    thymine=0
    cytosine=0
    for i in sequence:
        if i=='A'or i=="a":
            adenine+=1
        if i=='T'or i=="t":
            thymine+=1
        if i=='G'or i=="g":
            guanine+=1
        if i=='C'or i=="c":
            cytosine+=1
    per_a=(adenine/total_base_pairs)*100
    per_t=(thymine/total_base_pairs)*100
    per_g=(guanine/total_base_pairs)*100
    per_c=(cytosine/total_base_pairs)*100
    print("Percentage of A=",per_a,"\nPercentage of T=",per_t,"\nPercentage of G=",per_g,"\nPercentage of C=",per_c)
    return per_a, per_t, per_g, per_c
print("Nucleotide percentage of sequence 1:\n",percentage_basepairs(a))
print("Nucleotide percentage of sequence 2:\n",percentage_basepairs(b))

#This function is defined to trascribe DNA sequnces to RNA sequences and RNA sequence to Protein sequence.
def transcription_translation(dna):
    for ch in dna:
        rna=dna.replace("T","U")
    print("\nRNA sequence:",rna)
    
    #This dictionary contain the triplet codes as keys and their corresponding amino acids as their value
    codon={ "AAA":"Lys", "AAC":"Asn", "AAG":"Lys", "AAU":"Asn", 
            "ACA":"Thr", "ACC":"Thr", "ACG":"Thr", "ACU":"Thr", 
            "AGA":"Arg", "AGC":"Ser", "AGG":"Arg", "AGU":"Ser", 
            "AUA":"Ile", "AUC":"Ile", "AUG":"Met", "AUU":"Ile", 

            "CAA":"Gln", "CAC":"His", "CAG":"Gln", "CAU":"His", 
            "CCA":"Pro", "CCC":"Pro", "CCG":"Pro", "CCU":"Pro", 
            "CGA":"Arg", "CGC":"Arg", "CGG":"Arg", "CGU":"Arg", 
            "CUA":"Lys", "CUC":"Leu", "CUG":"Leu", "CUU":"Leu", 

            "GAA":"Glu", "GAC":"Asp", "GAG":"Glu", "GAU":"Asp", 
            "GCA":"Ala", "GCC":"Ala", "GCG":"Ala", "GCU":"Ala", 
            "GGA":"Gly", "GGC":"Gly", "GGG":"Gly", "GGU":"Gly", 
            "GUA":"Val", "GUC":"Val", "GUG":"Val", "GUU":"Val", 

            "UAA":"_", "UAC":"Tyr", "UAG":"_", "UAU":"Thr", 
            "UCA":"Ser", "UCC":"Ser", "UCG":"Ser", "UCU":"Ser", 
            "UGA":"_", "UGC":"Cys", "UGG":"Trp", "UGU":"Cys", 
            "UUA":"Leu", "UUC":"Phe", "UUG":"Leu", "UUU":"Phe"}
    word=[rna[i:i+3] for i in range(0, len(rna), 3)] 
    print("\nCodon in RNA: ",word)
    protein=""
    for triplet in word:
        if triplet in codon:
            protein=protein+codon[triplet]+'-'
    print("\nProtein sequence:",protein)
print("\n*******************************")
print("\nExon for sequence 1=",exon_a)
transcription_translation(exon_a)
print("\n*******************************")
print("\nExon for sequence 2=",exon_b)
transcription_translation(exon_b)
```
